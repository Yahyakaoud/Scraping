```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import random
import time

# Headers for requests (if needed)
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36'
}

# Initialize URL list and web driver
urls = []
driver = webdriver.Chrome()

# Main URL of Btech smartphones section
mainurl = 'https://btech.com/en/moblies/mobile-phones-smartphones/smartphones.html'
driver.get(mainurl)

# Scroll and load more content
while True:
    try:
        # Wait until the loading mask disappears
        WebDriverWait(driver, 10).until(
            EC.invisibility_of_element_located((By.CLASS_NAME, 'loading-mask'))
        )
        
        # Wait for and click the 'Show More' button
        try:
            show_more_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.CLASS_NAME, 'amscroll-load-button'))
            )
            time.sleep(10)  # A delay after button click
            show_more_button.click()
            print("Clicked 'Show More' button, loading more items...")
        except Exception:
            print("No more 'Show More' button found or unable to click. Exiting scroll loop.")
            break
        
        # Dynamic wait after clicking (wait for more content to load)
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, 'listingWrapperSection'))
        )
    except Exception as e:
        print(f"End of content or error: {e}")
        break

# Parse the final page source with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Extract all product URLs
url_tags = soup.find_all('a', class_='listingWrapperSection')
for url_tag in url_tags:
    urls.append(url_tag.get('href'))

# Close the browser
driver.quit()

# Output the number of URLs retrieved
urls = [url for url in urls if url is not None]
print(f'{len(urls)} URLs Retrieved')
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36'
}
titles, prices, warranties, rates, brands, OSs, rams, cpus = [], [], [], [], [], [], [], []
storages, SSs, resolutions, Display_Types, model_names, batteries, colors = [], [], [], [], [], [], []
maincameras, frontcameras, connectivities, NOSs, PMs, STs, Camera_Flashs = [], [], [], [], [], [], []
networks, usbs, touch, imgs, reviews,currancy,website,urls_view = [], [], [], [], [], [], [],[]
def scrape_btech_reviews(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36'}
    response = requests.get(url, headers=headers)
    
    if response.status_code != 200:
        return []
    
    soup = BeautifulSoup(response.content, 'html.parser')
    reviews_section = soup.find('div', class_='reviews-tab-content')
    if not reviews_section:
        return []

    reviews = reviews_section.find_all('div', class_='reviews-information-container')
    review_data = []

    for review in reviews:
        reviewer_name = review.find('div', class_='reviews-author')
        reviewer_name = reviewer_name.text.strip() if reviewer_name else 'N/A'
        review_rating = review.find('span', class_='star-count')
        review_rating = review_rating.text.strip() if review_rating else 'N/A'
        review_title = review.find('div', class_='reviews-title')
        review_title = review_title.text.strip() if review_title else 'N/A'
        review_date = review.find('div', class_='reviews-date')
        review_date = review_date.text.strip() if review_date else 'N/A'

        review_data.append({
            'reviewer_name': reviewer_name,
            'review_rating': review_rating,
            'review_title': review_title,
            'review_date': review_date
        })

    return review_data
for url in urls:
    try:
        PageToScrape = requests.get(url, headers=headers)
        if PageToScrape.status_code == 200:
            soup = BeautifulSoup(PageToScrape.text, 'html.parser')

            title_tag = soup.find('span', class_='base')
            price_tag = soup.find('div', class_='product-info-price')
            price_tag2 = price_tag.find('p', class_='price-huge-static')
            price = price_tag2.find('span',class_='hidden price-wrapper')
            price_val = price.find('span').text
            price = price_tag2.find('span',class_='hidden price-wrapper').text
            warranty_tag = soup.find('td', class_='col data warranty_val')
            maincamera_tag = soup.find('td', class_='col data main_camera_resolution_val')
            frontcamera_tag = soup.find('td', class_='col data front_camera_val')
            rate_tag = soup.find('div', class_='top-avg-rating-count')
            brand_tag = soup.find('td', class_='col data gfk_brand_16299_val')
            OS_tag = soup.find('td', class_='col data operating_system_type_val')
            ram_tag = soup.find('td', class_='col data ram_val')
            cpu_tag = soup.find('td', class_='col data number_of_processor_cores_val')
            storage_tag = soup.find('td', class_='col data storage_capacity_val')
            SS_tag = soup.find('td', class_='col data display_size_val')
            resolution_tag = soup.find('td', class_='col data resolution_val')
            Display_Type_tag = soup.find('td', class_='col data display_type_val')
            model_name_tag = soup.find('td', class_='col data model_name_val')
            Battery_tag = soup.find('td', class_='col data battery_capacity1_val')
            color_tag = soup.find('td', class_='col data color_val')
            NOS_tag = soup.find('td', class_='col data number_of_sim_val')
            PM_tag = soup.find('td', class_='col data chipset_manufacturer_val')
            connectivity_tag = soup.find('td', class_='col data connectivity_val')
            img_tag = soup.find('img', class_='no-sirv-lazy-load')
            ST_tag = soup.find('td', class_='col data sim_type_val')
            Camera_Flash_tag = soup.find('td', class_='col data camera_flash_val')
            network_tag = soup.find('td', class_='col data network_val')
            usb_tag = soup.find('td', class_='col data usb_type_val')
            touch_tag = soup.find('td', class_='col data touch_id_val')

            # Append data to lists with checks for None
            titles.append(title_tag.get_text(strip=True) if title_tag else None)
            prices.append(price_val if price_val else None)
            warranties.append(warranty_tag.get_text(strip=True) if warranty_tag else None)
            rates.append(rate_tag.get_text(strip=True) if rate_tag else None)
            brands.append(brand_tag.get_text(strip=True) if brand_tag else None)
            OSs.append(OS_tag.get_text(strip=True) if OS_tag else None)
            rams.append(ram_tag.get_text(strip=True) if ram_tag else None)
            cpus.append(cpu_tag.get_text(strip=True) if cpu_tag else None)
            storages.append(storage_tag.get_text(strip=True) if storage_tag else None)
            SSs.append(SS_tag.get_text(strip=True) if SS_tag else None)
            resolutions.append(resolution_tag.get_text(strip=True) if resolution_tag else None)
            Display_Types.append(Display_Type_tag.get_text(strip=True) if Display_Type_tag else None)
            model_names.append(model_name_tag.get_text(strip=True) if model_name_tag else None)
            batteries.append(Battery_tag.get_text(strip=True) if Battery_tag else None)
            colors.append(color_tag.get_text(strip=True) if color_tag else None)
            maincameras.append(maincamera_tag.get_text(strip=True) if maincamera_tag else None)
            frontcameras.append(frontcamera_tag.get_text(strip=True) if frontcamera_tag else None)
            connectivities.append(connectivity_tag.get_text(strip=True) if connectivity_tag else None)
            NOSs.append(NOS_tag.get_text(strip=True) if NOS_tag else None)
            PMs.append(PM_tag.get_text(strip=True) if PM_tag else None)
            STs.append(ST_tag.get_text(strip=True) if ST_tag else None)
            Camera_Flashs.append(Camera_Flash_tag.get_text(strip=True) if Camera_Flash_tag else None)
            networks.append(network_tag.get_text(strip=True) if network_tag else None)
            usbs.append(usb_tag.get_text(strip=True) if usb_tag else None)
            touch.append(touch_tag.get_text(strip=True) if touch_tag else None)
            imgs.append(img_tag.get('data-old-hires') if img_tag and img_tag.get('data-old-hires') else img_tag.get('src') if img_tag else None)
            reviews.append(scrape_btech_reviews(url))
            urls_view.append(url)
            currancy.append('EGP')
            website.append('Btech')

        else:
            print(f"Failed to retrieve {url}. Status code: {PageToScrape.status_code}")

    except Exception as e:
        print(f"An error occurred while processing {url}: {e}")
        titles.append(None)
        prices.append(None)
        warranties.append(None)
        rates.append(None)
        brands.append(None)
        OSs.append(None)
        rams.append(None)
        cpus.append(None)
        storages.append(None)
        SSs.append(None)
        resolutions.append(None)
        Display_Types.append(None)
        model_names.append(None)
        batteries.append(None)
        colors.append(None)
        maincameras.append(None)
        frontcameras.append(None)
        connectivities.append(None)
        NOSs.append(None)
        PMs.append(None)
        STs.append(None)
        Camera_Flashs.append(None)
        networks.append(None)
        usbs.append(None)
        touch.append(None)
        imgs.append(None)
        reviews.append(None)
        currancy.append(None)
        website.append(None)
        urls_view.append(url)

# Create DataFrame
df_btech = pd.DataFrame({
    'Title': titles,
    'Brand': brands,
    'OS': OSs,
    'Storage': storages,
    'RAM': rams,
    'CPU': cpus,
    'Processor Manufacturer': PMs,
    'Screen Size': SSs,
    'Main Camera': maincameras,
    'Front Camera': frontcameras,
    'Number Of SIM': NOSs,
    'SIM Type': STs,
    'Resolution': resolutions,
    'Display Type': Display_Types,
    'Connectivity': connectivities,
    'Camera Flash': Camera_Flashs,
    'Model': model_names,
    'Network': networks,
    'Battery Capacity': batteries,
    'Color': colors,
    'Price': prices,
    'Currancy': currancy,
    'Rate': rates,
    'IMG': imgs,
    'URL':urls_view,
    'Warranty': warranties,
    'Port Type': usbs,
    'Touch ID': touch,
    'Reviews': reviews,
    'Website' : website
})

df_btech
```

```python
import pandas as pd
import re
dfuu=['Octa Core', 'Hexa Core', 'Quad Core', 'Deca Core']
def find_cpu_in_title(title, cpu_sizes):
    if title:
        title_lower = title.lower()
        if 'cpu' in title_lower  :
            for cpu in cpu_sizes:
                if re.search(r'\b' + re.escape(cpu.lower()) + r'\b', title_lower, re.IGNORECASE):
                    return cpu
    return None
df_btech['CPU'] = df_btech.apply(lambda row: find_cpu_in_title(row['title'], dfuu) or row['cpu'], axis=1)
```

```python
import pandas as pd
import re

# List of RAM sizes for regex matching
allrams =["512GB", "256GB", "128GB", "64GB", "32GB", "16GB", "8GB"]

# Function to check if a RAM size is in the title using regex
def find_strg_in_title(title, storage_sizes):
    if title:
        title_lower = title.lower()
        # Check for keywords related to RAM
        
        if 'storage' in title_lower  :
            for storage in storage_sizes:
                if re.search(r'\b' + re.escape(storage.lower()) + r'\b', title_lower, re.IGNORECASE):
                    return storage
    return None
df_btech['Storage'] = df_btech.apply(lambda row: find_strg_in_title(row['title'], allrams) or row['storage'], axis=1)
```

```python
import pandas as pd
import re

# List of RAM sizes for regex matching
allrams = ["1GB", "2GB", "3GB", "4GB", "6GB", "8GB", "12GB"]

# Function to check if a RAM size is in the title using regex
def find_ram_in_title(title, ram_sizes):
    if title:
        title_lower = title.lower()
        # Check for keywords related to RAM
        if 'ram' in title_lower or 'memory' in title_lower:
            for ram in ram_sizes:
                if re.search(r'\b' + re.escape(ram.lower()) + r'\b', title_lower, re.IGNORECASE):
                    return ram
    return None
df_btech['RAM'] = df_btech.apply(lambda row: find_ram_in_title(row['title'], allrams) or row['ram'], axis=1)
```

```python
import pandas as pd
import re

# List of RAM sizes for regex matching

# Function to check if a RAM size is in the title using regex
def find_model_in_title(title, model_sizes):
    if title:
        title_lower = title.lower()
        # Check for keywords related to RAM
        
        if 'model' in title_lower  :
            for model in model_sizes:
                if re.search(r'\b' + re.escape(model.lower()) + r'\b', title_lower, re.IGNORECASE):
                    return model
    return None
df_btech['Model'] = df_btech.apply(lambda row: find_model_in_title(row['title'], dfm) or row['model'], axis=1)
```
